1. Handling Missing Data Questions

How do you identify and handle missing values:
We can use(      for identification       )
df.isnull() function to identify missing values in DataFrame, or
df.isnull().sum() to sum up number of missing values in respective columns. Also, we can use
round.((df.isnull().sum() / df.shape[0])*100, 2) to find out percentage of missing values in their respective columns

```python
print(fact.isnull())
```

as for handling, we can use:
df.drop(columns='Cabin') to remove the column depending the situation or
df.dropna(subset=[Embarked], axis=0, inplace=True) to remove the row

df.fillna(df.Age.mean()) can be used to impute(fill places of missing values with finding mean of whole column)

```python
print(fact.drop(columns='Country'))
```

```python
fact['Current account balance'] = fact['Current account balance'].fillna(fact['Current account balance'].mean())
print(fact)
```

"What is imputation, and why might it be useful in dealing with missing data?"
Imputation is one of the method to deal with missing data, it is filling missed data's sell by finding median or mean of that particular column

As for usefulnes of it:
=> imputations avoids losing data. In contrast to deletion method it doesn't delete a whole row or columns since there might useful info aside from missing data
=> it allows to make predictions by analizing already existing data.

2. Data Transformations Questions
"How can you encode catagorical variables in a Pandas DataFrame?"

We can encode them by using funtions like replace() or map() to numeric values:
=> .str.replace() - helps to replace strings with integers
=> .map({'string':1, ...}) - assigns integers as values for string keys

```python
fact2_1=fact
fact2_1['Country'].replace("Afghanistan", 1)
```

```python
fact2_2=fact
fact2_2['Country']=fact2_2['Country'].map({j: i for i, j in enumerate(fact2_2["Country"])})
print(fact2_2)
```
